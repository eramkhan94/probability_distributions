```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from scipy.stats import bernoulli, binom, geom, poisson
```

```python
# Set the style for seaborn
sns.set(style="whitegrid")

# Parameters
p_bernoulli = 0.5  # Probability for Bernoulli
n_binomial = 10     # Trials for Binomial
p_binomial = 0.5   # Probability for Binomial
p_geometric = 0.3  # Probability for Geometric
lambda_poisson = 5  # Lambda for Poisson

# Bernoulli Distribution
x_bernoulli = [0, 1]
y_bernoulli = bernoulli.pmf(x_bernoulli, p_bernoulli)

# Binomial Distribution
x_binomial = np.arange(0, n_binomial + 1)
y_binomial = binom.pmf(x_binomial, n_binomial, p_binomial)

# Geometric Distribution
x_geometric = np.arange(1, 11)
y_geometric = geom.pmf(x_geometric, p_geometric)

# Poisson Distribution
x_poisson = np.arange(0, 15)
y_poisson = poisson.pmf(x_poisson, lambda_poisson)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Bernoulli
axs[0, 0].bar(x_bernoulli, y_bernoulli, color='skyblue')
axs[0, 0].set_title('Bernoulli Distribution')
axs[0, 0].set_xticks(x_bernoulli)
axs[0, 0].set_ylim(0, 1)

# Binomial
axs[0, 1].bar(x_binomial, y_binomial, color='lightgreen')
axs[0, 1].set_title('Binomial Distribution (n=10, p=0.5)')
axs[0, 1].set_xticks(x_binomial)

# Geometric
axs[1, 0].bar(x_geometric, y_geometric, color='salmon')
axs[1, 0].set_title('Geometric Distribution (p=0.3)')
axs[1, 0].set_xticks(x_geometric)

# Poisson
axs[1, 1].bar(x_poisson, y_poisson, color='gold')
axs[1, 1].set_title('Poisson Distribution (Î»=5)')
axs[1, 1].set_xticks(x_poisson)

# Layout adjustment
plt.tight_layout()
plt.show()
```

```python

# Set up the figure and axis
x = np.linspace(-5, 5, 1000)

# Uniform distribution
uniform = stats.uniform.pdf(x, loc=-1, scale=2)  # Uniform between -1 and 1

# Exponential distribution
exponential = stats.expon.pdf(x, scale=1)  # Exponential with scale parameter 1

# Gamma distribution
gamma = stats.gamma.pdf(x, a=2, scale=1)  # Gamma with shape parameter 2

# Beta distribution
beta = stats.beta.pdf(x, a=2, b=5)  # Beta distribution with parameters 2 and 5

# Triangular distribution
triangular = stats.triang.pdf(x, c=0.5, loc=-1, scale=2)  # Triangular distribution

# Normal distribution
normal = stats.norm.pdf(x, loc=0, scale=1)  # Standard normal distribution

# Weibull distribution
weibull = stats.weibull_min.pdf(x, c=1.5)  # Weibull distribution with shape parameter 1.5

# Create subplots
fig, axs = plt.subplots(4, 2, figsize=(12, 16))

# Plot each distribution
axs[0, 0].plot(x, uniform, label='Uniform', color='blue')
axs[0, 0].set_title('Uniform Distribution')
axs[0, 0].legend()

axs[0, 1].plot(x, exponential, label='Exponential', color='orange')
axs[0, 1].set_title('Exponential Distribution')
axs[0, 1].legend()

axs[1, 0].plot(x, gamma, label='Gamma', color='green')
axs[1, 0].set_title('Gamma Distribution')
axs[1, 0].legend()

axs[1, 1].plot(x, beta, label='Beta', color='red')
axs[1, 1].set_title('Beta Distribution')
axs[1, 1].legend()

axs[2, 0].plot(x, triangular, label='Triangular', color='purple')
axs[2, 0].set_title('Triangular Distribution')
axs[2, 0].legend()

axs[2, 1].plot(x, normal, label='Normal', color='brown')
axs[2, 1].set_title('Normal Distribution')
axs[2, 1].legend()

axs[3, 0].plot(x, weibull, label='Weibull', color='cyan')
axs[3, 0].set_title('Weibull Distribution')
axs[3, 0].legend()

# Hide empty subplot
axs[3, 1].axis('off')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
